```python
def get_completion_from_messages(messages):
    response = client.chat.completions.create(
        model="sonar-pro",
        messages=messages,
        temperature=0,
        max_tokens=100,
    )
    return response.choices[0].message.content
# chat completion without streaming

def get_completion_from_messages_stream(messages):
    response = client.chat.completions.create(
        model="sonar-pro",
        messages=messages,
        temperature=0,
        max_tokens=100,
        stream=True,
    )
    for chunk in response:
        print(chunk.choices[0].delta.content, end="", flush=True)
    print()
```

```python
def get_process_info(pid=None, name=None, df_processes=None):
    """
    Get process information by PID or name.
    If both PID and name are provided, PID takes precedence.
    If name is provided, it searches for processes containing the name."""
    if pid is None:
        pid_values = df_processes[df_processes['Name'].str.contains(name, case=False, na=False)]['PID'].tolist()
    else:
        pid_values = [pid]
    
    return [get_process_details(p) for p in pid_values]

def get_process_details(pid):
    """
    Get detailed information about a process given its PID.
    Returns a dictionary with memory usage, CPU usage, and command line.
    """

    process = psutil.Process(pid)
    memory_info = process.memory_info()
    cpu_percent = process.cpu_percent(interval=1)
    cmd_line = process.cmdline()
    return {
        "memory_rss": memory_info.rss,
        "memory_vms": memory_info.vms,
        "cpu_percent": cpu_percent,
        "cmd_line": cmd_line,
    }.to_string()

def determine_processes_malicio
```

```python
import psutil
import pandas as pd
from datetime import datetime

# Initialize an empty list to store process information
process_list = []


# Iterate over all running processes
for proc in psutil.process_iter(['pid', 'name', 'status', 'cpu_percent', 'memory_percent', 'exe']):
    try:
        # Retrieve process data
        pid = proc.info['pid']
        name = proc.info['name']
        status = proc.info['status']
        cpu_usage = proc.info['cpu_percent']
        memory_usage = proc.info['memory_percent']
        exe_path = proc.info['exe']  # Process description
        # Additional process attributes
        open_files = proc.open_files()  # Open files
        is_running = proc.is_running()  # Running status
        parent_pid = proc.parent().pid if proc.parent() else None  # Parent process ID
        connections = proc.net_connections(kind='inet')  # Network connections
        create_time = proc.create_time()  # Process creation time

        # Prepare the process information dictionary
        process_info = {
            'PID': pid,
            'Name': name,
            'Status': status,
            'CPU Usage (%)': cpu_usage,
            'Memory Usage (%)': memory_usage,
            'Executable Path': exe_path,
            'Open Files': open_files,
            'Is Running': is_running,
            'Parent PID': parent_pid,
            'Connections': connections,
            'Creation Time': datetime.fromtimestamp(create_time).strftime('%m-%d-%Y %H:%M:%S') if create_time else None
        }

        # Append the process information to the list
        process_list.append(process_info)

    except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
        # Handle processes that are no longer running or inaccessible
        continue

# Convert the list of process information dictionaries into a pandas DataFrame
df_processes = pd.DataFrame(process_list)

# Display the DataFrame
display(df_processes)
```

```python
def get_running_processes():
    """Scan OS processes and return list of process names."""
    procs = []
    for p in psutil.process_iter(['pid', 'name']):
        try:
            procs.append(p.info['name'])
        except (psutil.NoSuchProcess, psutil.AccessDenied):
            continue
    return list(set(procs))  # Unique process names

get_running_processes()
```
